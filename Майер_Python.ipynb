```python
import numpy as np
```

```python
def month_exp(value):
    if pd.isna(value) or value == "Не указано":
        return np.nan 

    phrase = value.split('  ')[0]
    
    years = 0
    months = 0
    
    if "год" in phrase:
        years = int(phrase.split(" год")[0].split()[-1])
    elif "лет" in phrase:
        years = int(phrase.split(" лет")[0].split()[-1])
    if "месяц" in phrase:
        months = int(phrase.split("месяц")[0].split()[-1])
        
    return years * 12 + months
```

```python
df["Опыт работы (месяц)"] = df["Опыт работы"].apply(month_exp)
```

```python
import re

pattern = r"(готов(?:а)? к переезду|не готов(?:а)? к переезду|хочу переехать)"

def get_transfer(value):
    string = re.search(pattern, value).group(0)
    return not bool(re.search(r"\bне\b", string))
    
df["Готовность к переезду"] = df["Город, переезд, командировки"].apply(get_transfer)
```

```python
def get_travel(value):
    text = value.split(',')[-1]
    if 'командировк' not in text:
        return False
    if 'не' in text:
        return False
    return True
```

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
employment_categories = [
    "полная занятость", "частичная занятость", "проектная работа", "волонтерство", "стажировка"
]
schedule_categories = [
    "полный день", "сменный график", "гибкий график", "удаленная работа", "вахтовый метод"
]

for category in employment_categories:
    df[category] = df["Занятость"].apply(lambda x: category in x)

# Создаем "мигалки" для графика работы
for category in schedule_categories:
    df[category] = df["График"].apply(lambda x: category in x)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
df['дата_резюме'] = pd.to_datetime(df['Обновление резюме']).dt.date
```

```python
exdf['upd_date'] = pd.to_datetime(exdf['date']).dt.date
```

```python
exdf.head(1)
```

```python
def get_salary(salary):
    match = re.match(r'(\d+)\s*([^\d\s]+)', salary)
    currency_map = {
    'РУБ.': 'RUB',
    'БЕЛ.РУБ.': 'BYN',
    'KZT': 'KZT',
    'EUR': 'EUR',
    'USD': 'USD',
    'ГРН.': 'UAH',
    'СУМ': 'UZS',
    'KGS': 'KGS',
    'AZN': 'AZN'
    }
    if match:
        return int(match.group(1)), currency_map[match.group(2).upper()]
    return None, None
    
df[['сумма', 'валюта']] = df['ЗП'].apply(lambda x: pd.Series(get_salary(x)))
```

```python
df
```

```python
merged_df = pd.merge(
    df,
    exdf,
    left_on=['дата_резюме', 'валюта'],
    right_on=['upd_date', 'currency'],
    how='left'
)
```

```python
merged_df[merged_df.валюта != 'RUB']
```

```python
merged_df['close'] = merged_df['close'].fillna(1)
merged_df['proportion'] = merged_df['proportion'].fillna(1)
```

```python
df['ЗП (руб)'] = merged_df['сумма'] * merged_df['close'] / merged_df['proportion']
```
